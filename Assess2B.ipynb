###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
#d=d1.append(d2)
df = pd.concat(map(pd.read_csv, ['college_1.csv', 'college_2.csv']), ignore_index=True)
df
```

```python
#df['python_en'].fillna(0,inplace=True)
df.fillna(0,inplace=True)
df
```

###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


###### if  10000<codekata score<15000   (Reached_expectations.csv)



###### if  7000<codekata score<10000   (Needs_Improvement.csv)


###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
excedexpect=df[df['CodeKata Score'] >15000]
excedexpect.to_csv("ExceededExpectations.csv")
print(excedexpect)
```

```python
Reached_expec=df[(df['CodeKata Score']<15000) & (df['CodeKata Score']>10000)]
Reached_expec.to_csv("ReachedExpectations.csv")
print(Reached_expec)
```

```python
Needs_improv=df[(df['CodeKata Score']<10000) & (df['CodeKata Score']>7000)]
Needs_improv.to_csv("NeedsImprovement.csv")
print(Needs_improv)
```

```python
Unsat=df[df['CodeKata Score']<7000]
Unsat.to_csv("Unsatisfactory.csv")
print(Unsat)
```
